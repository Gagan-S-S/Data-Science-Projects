```python
import smtplib
import pandas as pd
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

# SMTP Configuration
SMTP_SERVER = "smtp.gmail.com"  # Update for your provider
SMTP_PORT = 587
SMTP_USER = "gaganshiremath973@gmail.com"  # Replace with your email
SMTP_PASS = "your-email-password"  # Use App Password for security

# Read email data from Excel
file_path = "emails.xlsx"  # Update with your file path
df = pd.read_excel(file_path)

# Ensure required columns exist
if "Email" not in df.columns:
    raise ValueError("Excel file must contain an 'Email' column.")

# Optional: Personalization
for index, row in df.iterrows():
    recipient_email = row["Email"]
    recipient_name = row.get("Name", "User")  # Use default if Name column is missing
    custom_message = row.get("Message", "Hello, this is a test email!")  # Default message
    
    # Construct Email
    subject = "Personalized Email"
    body = f"""\
    Hello {recipient_name},

    {custom_message}

    Best regards,
    Your Name
    """

    try:
        # Connect to SMTP Server
        server = smtplib.SMTP(SMTP_SERVER, SMTP_PORT)
        server.starttls()
        server.login(SMTP_USER, SMTP_PASS)

        # Create email
        msg = MIMEMultipart()
        msg["From"] = SMTP_USER
        msg["To"] = recipient_email
        msg["Subject"] = subject
        msg.attach(MIMEText(body, "plain"))

        # Send email
        server.sendmail(SMTP_USER, recipient_email, msg.as_string())
        print(f"Email sent to {recipient_email}")

        server.quit()
    
    except Exception as e:
        print(f"Error sending email to {recipient_email}: {e}")

print("All emails processed.")
```
